```python
def inList (my_char, my_list=['a','e','i','o','u']):
  """ this is a function to check whether a char is vowel or not """
  for each_char in my_list:
    if each_char == my_char.lower():
      return True
  return False
```

```python
inList('Z', ['x', 'y', 'z'])
```
