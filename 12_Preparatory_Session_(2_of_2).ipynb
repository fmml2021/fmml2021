## 2.3 Madagascar Problem

```python
import numpy as np
```

```python
A = np.array([[20, 122, 20],
              [35,180, 35],
              [50, 220, 65], 
              [70, 300, 30]])
x = np.array([[0.05], 
              [0.002], 
              [0.02]])
```

```python
# 4x3 . 3x1 = 4x1
```

```python
y = np.matmul(A,x) - 4.0
```

```python
y
```

```python
def highrisk(B, z):
  return (np.matmul(B,z) - 4.0)
```

```python
y = highrisk(A,x)
```

```python
Z = np.hstack((A,y))
```

```python
Z.astype(int)
```

```python
def RankArray(A,x):
  return np.hstack((A, (np.matmul(A,x)-4.0)))
```

```python
print (RankArray(A,x).astype(int))
```

## 2.4 Plotting with Matplotlib

#### Matplotlib is a multi-platform data visualization library built on NumPy arrays

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

```python
# create the first of two panels and set current axis
plt.subplot(1, 2, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x));

plt.show()
```

```python
# saving the figure

x = np.linspace(0, 10, 100)

fig = plt.figure()

plt.plot(x, np.sin(x), label='sine')
plt.plot(x, np.cos(x), label='cosine')


plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```
